# K-Fold

```python
# loading library
import os
import numpy as np
import pandas as pd
#For test train split
from sklearn.model_selection import train_test_split
#Loading a confusion matrix
from sklearn.metrics import confusion_matrix
from sklearn.metrics import mean_squared_error
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import KNeighborsRegressor
```

```python
os.chdir(r'C:\Users\Yogesh\Downloads')       #changing the directory
c = pd.read_csv('iris.csv')                  # loading data
c  = c.sample(150)                           #shuffling the data
c
```

```python
def Kclass(k,n):                                    # function Defnation             
    a = 0                                          # seting zero starting point
    b = int(len(c)/k)                              #converting to int & getiting len fro b 

    for x in range(k):                             #for loop starts
        v = c.iloc[a:b]                            # validation list starts
        v1= v.iloc[:,:4]                           #xTrain / valdidation data
        v2= v.iloc[:,4:5]                          #ytrain / valdidation data prediction
        u = c.iloc[0:a]                            #0 to a train data xtest 
        i = c.iloc[b:len(c)]                       #b to 150 train data xtest        
        a = b                                      #equating a and b for next iteraton
        b = int(b + len(c)/k)                      # find the len of b for last index
        frames = [u, i]                            #combine train data
        r = pd.concat(frames)                      #concate the x train data
        y1 =r.iloc[:,:4]                           #xTrain  
        y2 =r.iloc[:,4:5]                          #ytrain
        y2 = y2.squeeze()                          # to avoid warning
        knn = KNeighborsClassifier(n_neighbors=n)  #Kneighbors selected here
        knn.fit(y1 , y2)                           #fitting the model
        y_pred = knn.predict(v1)                   #getting the predited value
        conf_m= confusion_matrix(v2 , y_pred)      # confusion  matrix  
        print(conf_m)                              # printing confusion matrix 
        t = ((conf_m[0,0]+conf_m[1,1]+conf_m[2,2]) / np.sum(conf_m)) #getting the accuracy
        print(t)                                      # print the accuracy 
```

```python
Kclass(5,5)
```

# Regression

```python
os.chdir(r'C:\Users\Yogesh\Downloads')          #changing the directory
p = pd.read_csv('cars.csv')                     # loading the data
p  = p.sample(406)                              #shuffling the data
p
```

```python
def Kreg(k,n):
    a = 0                                      # seting a as zero starting point
    b = int(len(p)/k)                          #converting to int & getiting len fro b

    for x in range(k):                         #for loop starts          
        v = p.iloc[a:b]                        # validation list starts
        v1= v.iloc[:,3:7]                      #xTrain / valdidation data
        v2= v.iloc[:,1:2]                      #ytrain / valdidation data prediction
        u = p.iloc[0:a]                        #0 to a train data xtest 
        i = p.iloc[b:len(p)]                   #b to 150 train data xtest
        #i1 = i.iloc[:,4]        
        a = b                                  #equating a and b for next iteraton
        b = int(b + len(p)/k)                  # find the len of b for last index
        frames = [u, i]                        #combine train data
        r = pd.concat(frames)                  #concate the x train data
        y1 =r.iloc[:,3:7]                      #xTrain  
        y2 =r.iloc[:,1:2]                      #ytrain
        knn = KNeighborsRegressor(n_neighbors=n)#Kneighbors selected here 
        knn.fit(y1 , y2)                        #fitting the model
        y_pred = knn.predict(v1)                #getting the predited value
        mean=mean_squared_error(v2,y_pred)      #getting the squared error
        print(mean)                             #mean squared error
```
